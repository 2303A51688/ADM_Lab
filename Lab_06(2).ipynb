```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/university_student.csv')
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
numeric_df = df.select_dtypes(include=['number'])
```

```python
df_mean = numeric_df.mean()
print(df_mean)
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
from datetime import datetime
```

```python
df = pd.get_dummies(df, columns=["Gender", "School Type"], drop_first=True)
```

```python
df
```

```python
df["Grade"] = LabelEncoder().fit_transform(df["Grade"])
```

```python
df["Date of Birth"] = pd.to_datetime(df["Date of Birth"])
df["Age"] = datetime.now().year - df["Date of Birth"].dt.year
df["Total Score"] = df["Math Score"] + df["Reading Score"] + df["Writing Score"]
```

```python
df.drop(columns=["Date of Birth"], inplace=True)
```

```python
grouped_df = df.groupby("Grade").agg({
    "Total Score": "mean",
    "Age": "mean",
    "Absences": "sum"
}).reset_index()
```

```python
print("\nTransformed Dataset:")
print(df.head())

print("\nGrouped and Aggregated Data:")
print(grouped_df)
```
